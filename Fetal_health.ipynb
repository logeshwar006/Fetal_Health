# Feature Selection || Removing Unwanted Columns

# Importing necessary classification models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

# Splitting the dependent & independent variable X and y

```python
X=df.drop(columns="fetal_health")
y=df["fetal_health"]
```

```python
X.info()
```

```python
y.info()
```

# Category Encoder

```python
import category_encoders as ce
```

```python
encoder = ce.LeaveOneOutEncoder()
x=encoder.fit_transform(X,y)
```

```python
x.info()
```

# Importing metrices and train, test split 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Splitting the dataset for train and test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=40)
```

# Model Building

```python
# Decision Tree
decision_tree = DecisionTreeClassifier()

# Random Forest
random_forest = RandomForestClassifier()

# K-Nearest Neighbors
knn = KNeighborsClassifier()

# Support Vector Machine
svc = SVC()
```

```python
models=[decision_tree,random_forest,knn,svc]
```

```python
for model in models:
    model_name = model.__class__.__name__
    print(f"Fitting model: {model_name}")
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} accuracy: {accuracy}")
```
